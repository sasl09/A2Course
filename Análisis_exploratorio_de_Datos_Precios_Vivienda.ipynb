### 4.2.- Matriz de Correlación con Números

Ok, vamos haciendo lo mismo pero ahora con números - y vamos viendo quien se correlaciona más - sin llegar al punto que las 2 variables sean basicamente lo mismo claro

Comenzamos declarando k=10

```python
k = 10

mostCorrelated = corMat.nlargest(k, 'SalePrice') #This will return the 10 most corr vars (as rows) to SalePrice, but it
#will still show all other vars in df as cols
mostCorrelated
```

```python
k = 10

mostCorrelated = corMat.nlargest(k, 'SalePrice')['SalePrice'].index #This will return the 10 most corr to SalePrice only
mostCorrelated
```
